```python
# 1. Define paths 
s3_power_path = "s3://sarasota-power-data-bronze/power/"
checkpoint_path = "s3://sarasota-power-data-bronze/checkpoints/power/"
bronze_power_table = "default.bronze_power_data" 

# 2. Read the hourly JSON stream using Databricks Auto Loader
print("Pulling JSON files from S3...")
df_power = (spark.readStream
    .format("cloudFiles") 
    .option("cloudFiles.format", "json") 
    .option("cloudFiles.schemaLocation", checkpoint_path) 
    .load(s3_power_path))

# 3. Write to the Bronze Delta Table
(df_power.writeStream
    .format("delta") 
    .option("checkpointLocation", checkpoint_path)
    .trigger(availableNow=True) 
    .toTable(bronze_power_table))

print("Boom. Ingestion complete.")
```
